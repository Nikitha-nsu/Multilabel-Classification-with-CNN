```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
import keras
import pandas as pd
```

```python
#Create a CNN

#Import Libraries
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import Dense
from keras.layers import MaxPool2D
from keras.layers import Flatten
```

```python

# #Initializing CNN
classifier = Sequential()
```

```python

# #create first input layer #Convolution Operation
classifier.add(Conv2D(32, (3,3),activation = 'relu', input_shape = (64,64,3)))
classifier.add(MaxPool2D(pool_size=(2,2)))
```

```python
# #Adding Second Convolution layer
classifier.add(Conv2D(32, (3,3), activation = 'relu'))
classifier.add(MaxPool2D(pool_size=(2,2)))
```

```python
# #Flattening
classifier.add(Flatten())
```

```python
# #Adding Hidden layers
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units=6, activation='sigmoid'))
```

```python
# #Compiling CNN
classifier.compile(optimizer='adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
# #Data Preprocessing

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                    shear_range = 0.2,
                                    zoom_range = 0.2,
                                    horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

val_datagen = ImageDataGenerator(rescale = 1./255)

train_data = train_datagen.flow_from_directory('C:/Users/Nikitha/.spyder-py3/111880_269359_bundle_archive/seg_train/seg_train',
                                          target_size = (64, 64),
                                          batch_size = 32,
                                          class_mode = 'categorical')
test_data = test_datagen.flow_from_directory('C:/Users/Nikitha/.spyder-py3/111880_269359_bundle_archive/seg_test/seg_test',
                                          target_size=(64,64),
                                          batch_size = 32,
                                          class_mode = 'categorical')
val_data = val_datagen.flow_from_directory('C:/Users/Nikitha/.spyder-py3/111880_269359_bundle_archive/seg_pred/seg_pred',
                                          target_size=(64,64),
                                          batch_size = 32,
                                          shuffle=False, 
                                          class_mode = None) #Test with new data

```

```python
#steps_per_epoch(14034+3000)/32, #(total samples/batch size)
#validation_steps = 438 #(training samples/batch size)
```

```python
#Fit CNN to Images
classifier.fit_generator(train_data,
                         steps_per_epoch = 532,
                         epochs = 5,
                         validation_data = test_data,
                         validation_steps = 438
                         )
```

```python
#Perform prediction on new data
pred=classifier.predict_generator(val_data)
pred.shape
```

```python
pred_class_indices = np.argmax(pred,axis=1)
pred_class_indices
#Get ouput of index with max probabilities
```
